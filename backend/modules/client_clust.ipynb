```python
import pandas as pd

pd.options.display.max_rows = 999

pd.options.display.max_columns = 999
```

```python
df_facture = pd.read_csv('assets/dataframes/factures.csv')
df_produits = pd.read_csv('assets/dataframes/produits.csv')
df_produits_facture = pd.read_csv('assets/dataframes/produits_facture.csv')
df_users = pd.read_csv('assets/dataframes/users.csv')

df_facture
```

```python
df_users
```

```python
df = pd.DataFrame()

# reccuperation de l'id
df['id_utilisateur'] = df_users['id']
#reccuperation du code postal reduit
df['code_postal_short'] = df_users['postal_code'].astype(str).str[:2]
#reccuperation du code postal complet
df['code_postal_complet'] = df_users['postal_code']
#reccuperation de la ville
df['ville'] = df_users['city']
#reccuperation de létat
df['etat'] = df_users['state']
#reccuperation du genre
df['genre'] = df_users['genre']
df
```

```python
import pandas as pd
from dateutil.relativedelta import relativedelta
import datetime



# Recuperation de la collone date de naissance
df['date_naissance'] = pd.to_datetime(df_users['birth_day']).dt.date

# Reccuperation de lage a partir de la date de naissance
df['age'] = df['date_naissance'].apply(lambda x: relativedelta(datetime.date.today(), x).years)

df
```

```python
# reccuperation de la moyenne des total par facture. en gereral avec nous il depense combien
df['avg_total'] = (
    df_facture
    .groupby('user_id')['total']
    .transform('mean')
)

df
```

```python
#reccuperation du nombre de factures par utilisateur
df['nb_factures'] = df_facture.groupby('user_id')['facture_id'].transform('count')

#reccuperation du nombre de produits par utilisateur
df['total_produits'] = df_facture.groupby('user_id')['total_products'].transform('sum')
df['total_quantity'] = df_facture['total_quantities']

df
```

```python
df['total_spent'] = (
    df_facture
    .groupby('user_id')['total']
    .transform('sum')
)
```
